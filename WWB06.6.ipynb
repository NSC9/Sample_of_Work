```python
from sympy import *; x,h = symbols('x h')
```

```python
def f(x):
    return 9*x +7*sqrt(x)
```

```python
f(x)
```

```python
def g(x,h):
    return (f(x+h) - f(x))/h
limit(g(x,h),h,0)
```

```python
print(limit(g(x,h),h,0))
```

```python
plot(f(x))
```

```python
plot((limit(g(x,h),h,0)))
```
