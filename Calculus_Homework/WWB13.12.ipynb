```python
from IPython.display import Image
from IPython.core.display import HTML 
from sympy import *; x,h,y,t = symbols("x h y t")
Image(url= "https://i.imgur.com/UY7Fwag.png")
```

```python
# source https://nathancarter.github.io/how2data/site/how-to-find-the-critical-numbers-of-a-function-in-python-using-sympy/
f = -2*x**3 + 27*x**2 - 108*x+7
d = diff(f)
solve(Eq(d,0))
```

```python
p = plot(f,d, show = False, legend = True, )#xlim = (-20,20))
p[1].line_color = 'red'
p.show()
```
