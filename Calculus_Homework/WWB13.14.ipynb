```python
from IPython.display import Image
from IPython.core.display import HTML 
from sympy import *; x,h,y,t = symbols("x h y t")
Image(url= "https://i.imgur.com/sCApXGG.png")
```

```python
# source https://nathancarter.github.io/how2data/site/how-to-find-the-critical-numbers-of-a-function-in-python-using-sympy/
f = (4*x)/(8*x+2)
d = diff(f)
#solve(Eq(d,0)), 
Min(f,6).subs(x,2), Max(f,6).subs(x,2),f.subs(x,6)
```

```python
p = plot(f,d,show = False, legend = True, )#xlim = (-0.4,0.4), ylim = (-110,110))
p[1].line_color = 'red'
p.show()

```
