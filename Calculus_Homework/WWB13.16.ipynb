```python
from IPython.display import Image
from IPython.core.display import HTML 
from sympy import *; x,h,y,t = symbols("x h y t")
Image(url= "https://i.imgur.com/kEigUsc.png")
```

```python
# source https://nathancarter.github.io/how2data/site/how-to-find-the-critical-numbers-of-a-function-in-python-using-sympy/
f = 3*x**5 - 5*x**3 -5
d = diff(f)
f.subs(x,5)
```

```python
p = plot(f,d,show = False, legend = True, xlim = (-10,10), ylim = (-10,0))#xlim = (-0.4,0.4), ylim = (-110,110))
p[1].line_color = 'red'
p.show()

```
