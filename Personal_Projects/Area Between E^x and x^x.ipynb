```python
from sympy import E,plot,Eq,solve,Integral,diff
from sympy.abc import x
f = E**x
f
```

```python
g = x**x
g
```

```python
plot(x**x,E**x,yscale='log', xlim = (-1,3),ylim = (10**-0.2,10**1.3),size = (12,5),legend = True)
```

```python
# Step 1:
# Set f(x) = g(x) to find points of intersection
eq2 = Eq(f,g)
eq2
```

```python
solve(eq2)[0]
```

```python
# Step 2: Find the area between both curves
Integral(f-g,(x,0,E))
```

```python
Integral(f-g,(x,0,E)).evalf()
```

```python
# Now finding the area between their derivatives
plot(diff(x**x),diff(E**x),yscale='log', xlim = (-1,3),ylim = (10**-2,10**1.3),size = (12,5),legend = True)
```

```python
# Step 1:
# Set f'(x) = g'(x) to find points of intersection
eq3 = Eq(diff(f),diff(g))
eq3
```

```python
solve(eq3)[0]
```
