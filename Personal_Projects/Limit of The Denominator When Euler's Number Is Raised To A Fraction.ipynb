```python
from sympy import *; x,h,t,y,z,n   = symbols("x h t y z n", real=True)
f, g, h = symbols('f g h', cls=Function)

for n in range(1,21,5):   
    f = E**(x/n)
    g = E**(x/(n+1))
    h = E**(x/(n+2))
    p0 = plot(f,g,h,show = False,xlim = (1,10.5),size = (13,4),legend = True)
    p0[0].line_color = 'r'
    p0[1].line_color = 'g'
    p0[2].line_color = 'b'
    p0.show()
```
