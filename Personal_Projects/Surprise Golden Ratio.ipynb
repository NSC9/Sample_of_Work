```python
# Author - Github/NSC9
from sympy import *
from sympy.abc import x,p,n
```

```python
Integral(Derivative(x**n,x)-x**n,(x,0,Eq(x**n,Derivative(x**n,x))))
```

```python
for i in range(1,6):
    f = x**i
    eq1 = Eq(f,diff(f))
    b1 = int(nsolve(eq1,i))
    print("-------------------------------------------------------------------------------")
    print("Equations:  f' = ",diff(f),"    &    f=",f)
    print("Intersection of both is at x = ",b1)
    display(Eq(Integral(diff(f)-f,(x,0,b1)),Integral(diff(f)-f,(x,0,b1)).doit()))
    plot(f,diff(f), size=(10,4),xlim=(0,1.2*i),ylim=(0,(i**i+i**i)))
```

```python
g2 = x**(n+1)+x**n
Integral(Derivative(g2,x)-g2,(x,0,Eq(g2,Derivative(g2,x))))
```

```python
for k in range(1,6):
    g = x**(k+1)+x**k
    eq2 = Eq(g,diff(g))
    b2 = nsolve(eq2,2*k)
    print("-------------------------------------------------------------------------------")
    print("Equations:  g' = ",diff(g),"    &    g=",g)
    print("Intersection of both is at  x = ",b2 )
    display(Eq(Integral(diff(g)-g,(x,0,b2)),Integral(diff(g)-g,(x,0,b2)).doit(),evaluate=False)) 
    plot(g,diff(g), size=(10,4),xlim=(0,k+k),ylim=(0,(14*k**k)))
```
