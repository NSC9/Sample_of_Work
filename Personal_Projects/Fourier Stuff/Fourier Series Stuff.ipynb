```python


from sympy import symbols,E,fourier_series, pi,Derivative,Eq,plot
# Example

# Computing the Fourier series of f(x)=e**x at n = 3 , a = -pi, b = pi:
x = symbols('x')
f = E**x
f
```

```python
fs = fourier_series(f, (x, -pi, pi))
fst = fs.truncate(n=3)
fst
```

```python
# Compute the derivative of the Fourier series of e**x
e5 = Derivative(fst)
e6 = e5.doit()
Eq(e5,e6)
```

```python
u = 2
for i in ([u**q for q in range(8)]):
    p = plot(f,fs.truncate(n=i),Derivative(fs.truncate(n=i)).doit(),legend = True,xlim = (0,10),ylim = (-20,100),title="Analog Signal", xlabel = "Time(s)",ylabel = "Voltage (V)",size = (17,20),steps=1,show=False
        ) # here we are doing the same math above and plotting but with loops varying our n values of fourier equation
    p[0].line_color = 'r'
    p[1].line_color = 'b'
    p[1].label = 'fst(exp(x))'   # fst is fourie series truncated
    p[2].line_color = 'g'
    p[2].label = 'Derivative(fst(exp(x)))'
    p.show()
    print("                                                      i or n = ",i)
```
