```python
from sympy import *
from sympy.abc import x,y,n,z
f = Symbol('f_n')
eq = Sum(ceiling(cos(x)+n),(n, 0, oo))
display(Eq(f,eq))
display(Eq(Derivative(f),Derivative(eq).doit()))
display(Integral(f,(z,n,n+1)).evalf())
```

```python
def exhibit1(a,b):
    p = plot(show = False, size = (12,12), yscale= 'log')
    for n in range(a,b):
        sec =  ceiling(cos(x)+ n)
        p2 = plot(sec, show = False)
        p.append(p2[0])
    p.show()
exhibit1(0,100)
```
