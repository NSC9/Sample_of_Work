```python
from sympy import *
from sympy import euler
x = symbols('x')
c = symbols('c', real = True)
sympify(c).is_real
```

```python
eq5 = c*E**(x)
Eq(eq5,100) #solving for c when x = 10 because the plot looks nice
```

```python
plot(E**(x-5.4)) # after plotting this I noticed that it would be nice if this y= 100 right when x = 10
                 # the final solution may be useful for conceptualizing percentage of exponential growths
```

```python
eq11 = Eq(c*E**(10),100) 
eq11
# This is what I want to solve. 
```

```python
eq1 = 0.00453999208646404*exp(x)
Eq(eq1,eq1.subs(x,10).evalf())
```

```python
eq2 = 0.00454*exp(x)
Eq(eq2,eq2.subs(x,10).evalf())
```

```python
p0 = plot(eq1,eq2,show = False,xlim = (0,10),size = (12,5),legend = True,autoscale = True)
p0[0].line_color = 'r'
p0[1].line_color = 'b'
p0.show()
```

```python
s = solve(eq11)[0]
s # one possible solution for real number c when x = 10
```

```python
s.evalf()
```

```python
final_solution = (s*E**x)
phi = symbols('phi')
Eq(phi,final_solution)
```

```python
final_solution.subs(x,10)
```

```python
final_equation = s.evalf()*E**x
Eq(final_equation,100)
```
