```python
from sympy import *
from sympy.abc import x
from sympy import Poly
from random import randint, choice
%matplotlib widget
import matplotlib.cm as cm
from spb import *
import random

def generator():
    return Poly([choice((1,1))*randint(-10,10) for i in range(randint(2, 2))], x)

def stuff():

    P = generator().as_expr()
    a = str(P)[0:2]
    if a.find("*"):
        a = str(P)[0:1]
    root = solve(Eq(P,integrate(P,x)))
    A = min(root).evalf()
    B = max(root).evalf()
    eq = 1
    if integrate(P,x).replace(x,A/B) < P.replace(x,A/B):  # checks if b is negative. This swaps the integrands depending on b
        eq = -Integral(Integral(P,x)-P, (x,A,B))
    else:     
        eq = Integral(P-Integral(P,x), (x,A,B))  
    print("A = ",A)
    print("B = ",B)
    display(Eq(eq,eq.doit().evalf(),evaluate=False))
    plot(integrate(P), P, backend=MB,size = (9,4))
stuff()
```
