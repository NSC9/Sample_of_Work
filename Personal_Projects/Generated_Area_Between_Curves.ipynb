```python
from sympy import *
from sympy.abc import x
from sympy import Poly
from random import randint, choice
%matplotlib widget
import matplotlib.cm as cm
from spb import *
P = Poly([choice((-1,1))*randint(1,10) for i in range(randint(2, 2))], x).as_expr()
eq2 = Eq(P, integrate(P))
A = solve(eq2)[0]
B = solve(eq2)[1]
if B>A:
    eq = Integral(P-Integral(P), (x,A,B))
else: eq = Integral(P-Integral(P), (x,B,A))
display(Eq(eq,eq.doit().evalf()))
plot(integrate(P), P, backend=MB,size = (9,4))
```
