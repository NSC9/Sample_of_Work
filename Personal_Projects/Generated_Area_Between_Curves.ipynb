```python
from sympy import *
from sympy.abc import x
from sympy import Poly
from random import randint, choice
%matplotlib widget
import matplotlib.cm as cm
from spb import *
import random
Pp = Poly([choice((1,1))*randint(1,10) for i in range(randint(2, 2))], x).as_expr()
Pn = Poly([choice((-1,-1))*randint(1,10) for i in range(randint(2, 2))], x).as_expr()
eqp = Eq(P, integrate(P))
eqn = Eq(P, integrate(P))
Ap = solve(eqp)[0]
Bp = solve(eqp)[1]
An= solve(eqn)[0]
Bn = solve(eqn)[1]
if Bp>Ap:
    eq1 = Integral(Pp-Integral(Pp), (x,A,B))
else: 
    eq1 = Integral(Pp-Integral(Pp), (x,B,A))
if Bn>An:
    eq2 = Integral(Integral(Pn)-Pn, (x,B,A))
else: 
    eq2 = Integral(Integral(Pn)-Pn, (x,A,B))
i = 0
def random_value2(i):
    return random.randint(1, 2)
if random_value2(i) > 1:
    display(Eq(eq1,eq1.doit().evalf()))
    plot(integrate(Pp), Pp, backend=MB,size = (9,4))
else: 
    display(Eq(eq2,eq2.doit().evalf()))
    plot(integrate(Pn), Pn, backend=MB,size = (9,4))
```
