```python
from sympy import *
from sympy.abc import x
from sympy import Poly
from random import randint, choice
%matplotlib widget
import matplotlib.cm as cm
from spb import *
import random

def generator():
    b = choice((1,1))*randint(-10,10)
    P = Poly([b for i in range(randint(2, 2))], x).as_expr()
    root = solve(Eq(P,integrate(P,x)))
    A = min(root).evalf()
    B = max(root).evalf()
    eq = 1
    if b < 0:  # checks if b is negative. This swaps the integrands depending on b
        eq = -Integral(Integral(P,x)-P, (x,A,B))
    elif b > 0:     
        eq = Integral(P-Integral(P,x), (x,A,B))
    print("b =",b)    
    print("A = ",A)
    print("B = ",B)
    display(Eq(eq,eq.doit().evalf(),evaluate=False))
    plot(integrate(P), P, backend=MB,size = (9,4))
generator()
```
