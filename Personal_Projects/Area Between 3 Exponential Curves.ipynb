```python
from sympy import E,plot,Eq,nsolve,Integral,diff,Piecewise
from sympy.abc import x
f = E**x
f
```

```python
g = x**x
g
```

```python
plot(f,g,yscale='log', xlim = (-1,3),ylim = (10**-0.2,10**1.3),legend = True, size = (12,5))
```

```python
# Step 1:
# Set f(x) = g(x) and solve to find points of intersection
eq2 = Eq(f,g)
eq2
```

```python
# nsolve finds where both functions intersect
b1 = nsolve(eq2,2.7)
b1 
```

```python
# Step 2: Find the area between both curves
Integral(f-g,(x,0,b1))
```

```python
total = Integral(f-g,(x,0,b1)).evalf()
total
```

```python
# Now finding the area between the differentiated curves
plot(diff(f),diff(g),yscale='log', xlim = (-1,3),ylim = (10**-1.3,10**1.3),legend = True, size = (11,5))
```

```python
# Step 1:
# Set f'(x) = g'(x) to find points of intersection
eq3 = Eq(diff(f),diff(g))
eq3
```

```python
# this is our upper bound for our upcoming Integral
b2 = nsolve(eq3,2)
b2
```

```python
# Step 2: Find the area between both curves
Integral(diff(f)-diff(g),(x,0,b2))
```

```python
Integral(diff(f)-diff(g),(x,0,b2)).evalf(10)
```

```python
###################################################################################################################
```

```python
# Finding the area for the left half between these 3 curves
plot(f,g,diff(g),yscale='log', xlim = (-1,3),ylim = (10**-0.2,10**1.3),legend = True,size = (12,5))
```

```python
Integral(f-g,(x,0,b1))
```

```python
eq4 = Eq(g,diff(g))
nsolve(eq4,1)
```

```python
p = Piecewise((g, x < 1), (diff(g), x > 1))
p
```

```python
Integral(f-p,(x,0,b2))
```

```python
left_half = Integral(f-p,(x,0,b2)).evalf()
left_half
```

```python
right_half = total-left_half
right_half
```
