```python
%matplotlib widget
from sympy import *
from sympy.abc import x
p1 = plot(show=False,legend=False)
for n in range(1, 20):
    p2 = plot(integrate(cos(x)**n,x), show=False)
    p3 = plot(integrate(-cos(x)**n,x), show=False)   
    p1.append(p2[0])
    p1.append(p3[0])
p1.show()
```
