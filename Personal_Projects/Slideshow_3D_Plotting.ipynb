```python
from sympy import *         # import sympy library into our development environment
x = symbols('x')            # x  becomes a non-type sympy symbols that can be substited for python objects later
f = sin(x)**tan(x)          # storing our function as variable f
f                           # outputting our function in LaTex
```

```python
from sympy.plotting import plot3d
for i in range(0,6,1):       # range(start, finish, steps)
    if i != 0:
        plot3d(f, (x, 0, i), (y, 0, i), title = i)   
```

```python
for i in range(0,6,1):       # range(start, finish, steps)
    if i != 0:
        plot3d(-f, (x, 0, i), (y, 0, i), title = i)   # inversed the loot by making our function negative
```
