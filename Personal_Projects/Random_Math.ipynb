```python
from sympy import sin,cos,tan,cot,sec,csc,sinc,asin,acos,atan,acot,asec,acsc,sinh,cosh,tanh,LambertW,diff
from sympy import coth,sech,csch,asinh,acosh,atanh,acoth,asech,acsch,ceiling,floor,exp,sqrt,plot,log
from sympy.abc import x
import random
```

```python
funs = [sin(x),cos(x),tan(x),cot(x),sec(x),csc(x),sinc(x),abs(x),
            asin(x),acos(x),atan(x),acot(x),asec(x),acsc(x),sinh(x),
            cosh(x),tanh(x),coth(x),sech(x),csch(x),asinh(x),acosh(x),atanh(x),
            acoth(x),asech(x),acsch(x),ceiling(x),floor(x),exp(x),sqrt(x),
            log(x)]
for i in range(0,10):
    def random_math(i):  
        return funs[random.randrange(0,len(funs))]
    eq = random_math(i)*random_math(i)
    display(eq)
    plot(eq**-1,eq,(eq+1),(eq+1)**-1,(eq+2),(eq+2)**-1,(eq+3),(eq+3)**-1,(eq+4),(eq+4)**-1,(eq+5),(eq+5)**-1,
         (eq+6),(eq+6)**-1,(eq+7),(eq+7)**-1,(eq+8),(eq+8)**-1,(eq+9),(eq+9)**-1,(eq+10),(eq+10)**-1,
         (eq+11),(eq+11)**-1,(eq+12),(eq+12)**-1,(eq+13),(eq+13)**-1,(eq+14),(eq+14)**-1,
         size = (13,7),legend=False,yscale='log')
```
