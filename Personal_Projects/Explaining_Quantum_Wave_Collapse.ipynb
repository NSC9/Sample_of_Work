```python
from sympy import *
from sympy.abc import x
from sympy import factor_terms
delta= symbols('delta', real = True, constant = True)
f = Function("f")
a = delta*f(x)
print("Functions we use:")
display(Eq(1,a))
display(Eq(2,a**-1))
display(Eq(3,factor_terms(Integral(a,x))))
display(Eq(4,factor_terms(Integral(a,x)**-1)))
display(Eq(5,Derivative(a,x).doit()))
display(Eq(6,Derivative(a,x).doit()**-1))
display(Eq(7,gamma(a)))
display(Eq(8,gamma(a)**-1))
display(Eq(9,Limit(gamma(a),delta,(-oo,oo))))
```

```python
def plotter(c,d,s):
    for delta in range(c,d,s):
        eq = delta*cos(x)
        plot(eq,eq**-1,integrate(eq),integrate(eq)**-1,diff(eq),diff(eq)**-1,gamma(eq),gamma(eq)**-1,
             yscale='log', size=(14,5),legend=True, xlim = (-3,3))
plotter(10,11,1)
```

```python
plotter(100,101,1)
```
