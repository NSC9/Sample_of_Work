```python
from sympy import *
from sympy.abc import x
from sympy import Poly
from random import randint, choice
%matplotlib widget
import matplotlib.cm as cm
from spb import *
```

```python
def integrator():
    P = Poly([choice((1,1))*randint(-10,10) for i in range(randint(2, 2))], x).as_expr()
    display(P)
    p1 = plot(show=False,yscale="log",size=(7,7),backend=MB)
    for i in range(1,10):
        display(integrate(P,x))
        p2 = plot(integrate(P,x)**i, show=False,label=None,backend=MB)
        p3 = plot(integrate(P,x)**-i, show=False,label=None,backend=MB) 
        p1.append(p2[0])
        p1.append(p3[0])
        P = integrate(P,x)
    p1.show()
integrator()
```
