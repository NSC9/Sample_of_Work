```python
from sympy import *; x,h = symbols('x h')
```

```python
def f(x):
    return 2*x**3+3*x**2-120*x+5
f(x)
```

```python
plot(f(x))
```

```python
f(x.subs(x,-5))
```

```python
f(x.subs(x,3.5))
```

```python
def g(x,h):
    return (f(x+h)-f(x))/h
g(x,h)
```

```python
limit(g(x,h),h,0)
```

```python
print(limit(g(x,h),h,0))
```

```python
plot(limit(g(x,h),h,0))
```

```python
solve(limit(g(x,h),h,0))
```
